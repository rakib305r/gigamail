```python
import matplotlib.pyplot as plt

x = [1,2,3,4]
y = [10,20,30,40]

plt.plot(x,y) 
plt.title('Graph')
plt.xlabel('x axis') # Line Chart with Title + Labels
plt.ylabel('y axis')
plt.show()
```

```python
food = ['Mango', 'Banana', 'Apple', 'Berry']
rate = [8, 9, 7, 4 ] # Bar Chart 
plt.bar(food, rate)
plt.xlabel("food")
plt.ylabel("rate")

plt.show()
```

```python
x = [5,10,15,20,25,30]
y = [100,800,400,900,200,300]
plt.scatter(x,y)
plt.xlabel('x axis')  # Scatter Plot
plt.ylabel('y axis')

plt.show()
```

```python
data = [10,20,20,30,30,30,40,50,50,60,70]

plt.hist(data)
plt.title("Histogram Example") # Histogram
plt.xlabel("Value")
plt.ylabel("Frequency")

plt.show()
```

```python
x = [5,10,15,20,25,30]
y = [100,800,400,900,200,300]
plt.scatter(x,y, color='red')
plt.title("Age vs Salary Relation") # Scatter Plot
plt.xlabel("Age")
plt.ylabel('Salary')

plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot([5,15,10,30,25],[10,50,40,35,25])
plt.title("Line plot")

plt.subplot(1,2,2)
plt.bar(['Rakib','mitu', 'karim', 'juyel'],[25,50,10,30]) # Subplot (Multiple Graph)
plt.title("Bar Plot")

plt.show()
```

```python
x = [5,10,15,20,25,30]
y = [100,800,400,900,200,300]
plt.plot(x, y, color="orange", label="Growth line") # Professional Styling signaficantly grid and legend
plt.title("Styled Graph")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")

plt.grid(True)
plt.legend()

plt.show()
```
