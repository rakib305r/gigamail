```python
import seaborn as sns
import matplotlib.pyplot as plt
x =[1,2,3,4,5]
y = [10,20,30,40,50]
sns.lineplot(x = x ,y = y) # Simple Line Plot
plt.title("Asia")

plt.show()
```

```python
sns.scatterplot(x=x , y=y)
plt.title("Urupe")   # Scatter Plot

plt.show()
```

```python
subject = ['Math', 'English', 'Science', 'Religion']
marks = [80, 98, 88, 90]
sns.barplot(x=subject, y=marks) # Bar Plot
plt.title("Australia")

plt.show()
```

```python
data = [10,20,20,30,30,30,40,50,50,60]
sns.histplot(data)
plt.title("Bangladesh") # Histogram
```

```python
data = [10,50,20,30,15,30,40,50,500]
sns.boxplot(x = data)
plt.title("Box Plot") # Box Plot (Outlier Detect)

plt.show()
```

```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.DataFrame({
    "Age": [18, 20, 25, 30],
    "Salary": [2000, 5000, 6000, 8000] # Heatmap
})
sns.heatmap(df.corr(), annot=True) # annot er kaj box er bhitor er correlation value dekhanu
plt.show()
```

```python
tips = sns.load_dataset("tips") # Seaborn কিছু ready dataset দেয় practice করার জন্য: tips 🍽️ (restaurant data) ,iris 🌸 (flower data)
sns.pairplot(tips) # Pairplot (Magic Plot ✨)
plt.show()
```
